```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import csv

with open("Diwan.csv","w",encoding="UTF-8",newline='') as fd:
    heads = ["Book Name","Price"]
    writer = csv.DictWriter(fd,heads)
    writer.writeheader()
    
    for i in range(1,150):
        resp = requests.get(f"https://diwanegypt.com/product-category/books/arabic-books/page/{str(i)}/")
        src = resp.content
        soup = BeautifulSoup(src,"lxml")
        all_books = soup.find_all("li",{"class":"product"})

        for book in all_books:
            book_title = book.find("h2").text
            book_price = book.find("bdi").text
            writer.writerow({"Book Name":book_title,"Price":book_price})
```
