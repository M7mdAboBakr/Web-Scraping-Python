```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import csv

with open("WUZZUF.csv","w",newline='',encoding="UTF-8") as fd:

    heads = ['Title',"Location","Company Name","Experience","Skills"]
    writer = csv.DictWriter(fd,heads)
    writer.writeheader()

    for i in range(21):


        url = f"https://wuzzuf.net/search/jobs/?a=spbg&q=data%20scientist%20data%20analyst&start={i}"

        resp = requests.get(url)

        src = resp.content

        soup = BeautifulSoup(src,"lxml")

        all_jobs = soup.find_all("div",{"class":"css-1gatmva e1v1l3u10"})

        for job in all_jobs:

            job_title = job.find("h2").get_text()
            
            location = job.find("span",{"class":"css-5wys0k"}).get_text().strip()

            company_name = job.find("a",{"class":"css-17s97q8"}).get_text().strip('- ')

            experience = job.find_all("a",{"class":"css-o171kl"})[1].get_text()

            skills = job.find_all("a",{"class":"css-o171kl"})[1].find_next_siblings()

            Skills = []
            for skill in skills:

                if skill.get_text().strip().startswith(".css"):
                    continue
                
                Skills.append(skill.get_text().strip()[2:])
                
            writer.writerow({"Title":job_title,"Location":location,"Company Name":company_name,"Experience":experience,"Skills":Skills})

```
